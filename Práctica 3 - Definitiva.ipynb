# 1. Introducción

El objetivo de esta práctica es realizar un análisis completo de un problema de mantenimiento predictivo industrial, desde la exploración inicial hasta la generación de predicciones, aplicando modelos supervisados como árboles de decisión y Naive Bayes.

Se proporciona un conjunto de datos sintético que simula lecturas de sensores en condiciones industriales reales, incluyendo variables numéricas, categóricas y posibles patrones temporales. A diferencia de prácticas anteriores, no se especifica el contexto exacto, fomentando una interpretación autónoma de los datos y la toma de decisiones basada en evidencia.

El proceso incluye: carga, exploración, preprocesamiento, entrenamiento, evaluación y predicción sobre un conjunto de test. La evaluación final se realizará mediante una competición en Kaggle, usando el `F1-score` como métrica principal.

### 1.1. Datos

Se entregan tres archivos: 
- `train.csv`: Observaciones etiquetadas, con múltiples variables técnicas y la variable objetivo `fallo`, binaria (valores `1` (fallo) o `0` (normal)) y desbalanceada (alrededor de un 25% de fallos).
- `test.csv`: Observaciones sin etiqueta para evaluación.
- `sample_submission.csv`: Ejemplo de cómo hay que formatear la solución para subir a la competición de Kaggle.

Los datos deben tratarse como una caja negra: no se da información adicional sobre su origen. Se espera que el análisis determine el tratamiento más adecuado para cada variable, ya que pueden contener artefactos realistas como valores faltantes o ruido.

### 1.2. Desarrollo

El flujo de trabajo seguirá, de forma general, las siguientes etapas:

- **Análisis exploratorio** (*EDA - Exploratory Data Analysis*): distribución de la variable objetivo, comportamiento de las variables, relaciones entre ellas... identificando patrones, relaciones e irregularidades.

  > A veces es necesario incluir en esta sección alguna parte del preprocesamiento (por ejemplo, la creación de una variable nueva) para analizarla gráficamente.

- **Preprocesamiento** (*Feature Engineering, Data Cleaning...*): tratar los valores faltantes, codificar variables categóricas (*ej. one-hot*) o numéricas (*ej. discretización*), normalización si es necesaria, ...

- **Modelización**: al menos dos algoritmos de clasificación, comparando su rendimiento.

- **Predicción y evaluación**: aplicar el mejor modelo al conjunto de test y generar predicciones. Participación en una competición privada en Kaggle, usando `F1-score` como métrica principal. El objetivo es maximizar el `F1-score`, que es especialmente relevante debido al desbalance en los datos.

### **1. Estudio variable objetivo**

```python
f,ax=plt.subplots(1,2,figsize=(18,8))
df_train['fallo'].value_counts().plot.pie(explode=[0,0.1],autopct='%1.1f%%',ax=ax[0],shadow=True)
ax[0].set_title('fallo')
ax[0].set_ylabel('')
sns.countplot(x='fallo',data=df_train,ax=ax[1])
ax[1].set_title('fallo')
plt.show()
```

En ambas gráficas observamos que la variable fallo está desbalanceada, con una distribución de 76,3% de la clase 0 (sin fallo) y un 23,7% de la clase 1 (con fallo).

Este desbalance implica que, al analizar o modelar, no basta con maximizar la exactitud global, sino que conviene emplear métricas y técnicas específicas, como el **F1-score**.

### **2. Estudio variables categóricas**

#### 2.1 Estudio variable modo de operación

```python
df_train.groupby(['modo_operacion','fallo'])['fallo'].count()
```

```python
f,ax=plt.subplots(1,2,figsize=(18,8))
df_train[['modo_operacion','fallo']].groupby(['modo_operacion']).mean().plot.bar(ax=ax[0])
ax[0].set_title('fallo vs modo_operacion')
sns.countplot(x='modo_operacion',hue='fallo',data=df_train,ax=ax[1])
ax[1].set_title('modo_operacion: fallo vs acierto')
plt.show()
```

Con estas dos gráficas analizamos la relación entre la variable modo_operación y fallo, y observamos cómo el modo de operación condiciona el comportamiento de los fallos.
- Gráfica de fallo vs modo_operación:
    - Auto: Es la que menor tasa de fallos tiene y el modo más fiable, con sólo un 14% de fallos.
    - Mantenimiento: Presenta alrededor de 26% de fallo.
    - Manual: Presenta la tasa de fallos más alta, alrededor de un 39% de fallo.

- Gráfica modo_operación: fallo vs acierto:
    - Auto: Acumula alrededor de 1200 fallos debido a su elevado uso.
    - Mantenimiento: Alrededor de 400 fallos.
    - Manual: Genera 2000 fallos, siendo el mayor contribuyente al total de errores. 

Con estos datos, podemos concluir que el modo de operación influye decisivamente en el comportamiento de los fallos.

Estas gráficas muestran el comportamiento de la variable operador.
 
En cuanto al volumen de operaciones, el operador E registra el mayor número de operaciones, seguido de B, F, D y C respectivamente.

Por otro lado, la que mayor tasa de fallos presenta es el operador E, mientras que las de menor fallos son D y B, siendo esta última la de menor tasa de error.

```python
pd.crosstab([df_train.modo_operacion,df_train.fallo],df_train.operador,margins=True).style.background_gradient(cmap='summer_r')
```

### **5. Estudio variable temporal**

#### 5.1 Variable Año

```python
g = sns.catplot(
    data=df_train,
    x='operador', y='fallo',
    hue='modo_operacion',
    col='year',
    kind='bar',
    height=4, aspect=1,
    errorbar=None,
)

g.fig.subplots_adjust(top=0.85)
g.fig.suptitle("Tasa de fallo por operador y modo de operación, separado por año")
plt.show()
```

Estas gráficas confirman que, año tras año, el operador B y el modo auto mantienen las tasas de fallo más bajas, mientras que el modo manual y el operador E siguen siendo los puntos críticos.

# 3. Preprocesamiento

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import KBinsDiscretizer
from sklearn.preprocessing import StandardScaler
```

### 3.1 Train

```python
df_train = pd.read_csv('train.csv')
```

```python
df_train["timestamp"] = pd.to_datetime(df_train["timestamp"])

df_train["month"] = df_train["timestamp"].dt.month
df_train["hour"] = df_train["timestamp"].dt.hour
df_train["dayofweek"] = df_train["timestamp"].dt.dayofweek
df_train["dayofweek"] = df_train["dayofweek"].map({0: "Lunes", 1: "Martes", 2: "Miercoles", 3: "Jueves", 4: "Viernes", 5: "Sabado", 6: "Domingo"})
df_train['is_weekend'] = (df_train['timestamp'].dt.dayofweek >= 5).astype(int)
```

```python
df_train['operador_missing'] = df_train['operador'].isna().astype(int)
df_train["operador"] = df_train["operador"].fillna("B")

df_train['presion_missing'] = df_train['presion'].isna().astype(int)
imputer = KNNImputer(n_neighbors=3)
variables_relacionadas = ["presion",'sensor_3','temperatura','hour']
df_train[variables_relacionadas] = imputer.fit_transform(df_train[variables_relacionadas])
```

Para el tratamiento de los valores nulos:

- **Presión**. Empleamos **KNNImputer** para rellenar los valores nulos en presión. Cada valor se estima con la media de sus tres vecinos más similares en el espacio definido por las variables sensor_3, temperatura y hour, que muestran la mayor correlación con presión. 

Además, hemos añadido dos variables **"operador_missing"** y **"presión_missing"** para identificar donde faltan datos.

```python
df_test = pd.read_csv('test.csv')
df_test_id = df_test[['id']].copy()
```

```python
df_test["timestamp"] = pd.to_datetime(df_test["timestamp"])

df_test["month"] = df_test["timestamp"].dt.month
df_test["hour"] = df_test["timestamp"].dt.hour
df_test['dayofweek'] = df_test["timestamp"].dt.dayofweek
df_test["dayofweek"] = df_test["dayofweek"].map({0: "Lunes", 1: "Martes", 2: "Miercoles", 3: "Jueves", 4: "Viernes", 5: "Sabado", 6: "Domingo"})
df_test['is_weekend'] = (df_test['timestamp'].dt.dayofweek >= 5).astype(int)
```

```python
df_test["operador_missing"] = df_test["operador"].isna().astype(int)
df_test["operador"] = df_test["operador"].fillna("B")


df_test["presion_missing"] = df_test["presion"].isna().astype(int)
imputer = KNNImputer(n_neighbors=3)
variables_relacionadas = ["presion",'sensor_3','temperatura','hour']
df_test[variables_relacionadas] = imputer.fit_transform(df_test[variables_relacionadas])
```

```python
cont_vars = ["temperatura","presion","sensor_3"]
cat_vars = ["modo_operacion","operador","month","hour","is_weekend","dayofweek","presion_missing","operador_missing"]  

cont_vars_idx = [df_train.columns.get_loc(var) for var in cont_vars]
cat_vars_idx = [df_train.drop(columns=['fallo']).columns.get_loc(var) for var in cat_vars]

all_vars = ["modo_operacion","operador","temperatura","presion","sensor_3","month","hour","is_weekend","dayofweek","presion_missing","operador_missing"]
```

```python
X_train, X_test = train_test_split(df_train,test_size=0.2,random_state=532)
y_train, y_test = X_train["fallo"], X_test["fallo"]
X_train = X_train.drop(columns=["fallo"],axis=1)
X_test = X_test.drop(columns=["fallo"],axis=1)
```

El método train_test_split divide el conjunto de datos Train en dos subconjuntos:
- X_train: Contiene el 80 % para entrenar el modelo.
- X_test: Contiene el 20 % restante para evaluar su rendimiento.

```python
X_train_onehot = pd.get_dummies(X_train, columns=cat_vars)
X_test_onehot = pd.get_dummies(X_test, columns=cat_vars)

X_train_onehot, X_test_onehot = X_train_onehot.align(X_test_onehot, join='outer', axis=1, fill_value=0)
X_train_onehot, X_test_onehot = X_train_onehot.to_numpy(), X_test_onehot.to_numpy()

X_train = X_train.to_numpy()
X_test = X_test.to_numpy()

X_train_cont = X_train[:, cont_vars_idx]
X_test_cont = X_test[:, cont_vars_idx] 
```

```python
scaler = StandardScaler()

X_train_scaled_onehot = scaler.fit_transform(X_train_onehot)
X_test_scaled_onehot = scaler.transform(X_test_onehot)
```

```python
enc = OrdinalEncoder()
X_train[:, cat_vars_idx] = enc.fit_transform(X_train[:, cat_vars_idx])
X_test[:, cat_vars_idx] = enc.fit_transform(X_test[:, cat_vars_idx])

X_train_cat = X_train[:, cat_vars_idx]
X_test_cat = X_test[:, cat_vars_idx]

discretizer = KBinsDiscretizer(n_bins=5, encode='ordinal', strategy='uniform')
X_train_disc = discretizer.fit_transform(X_train_cont)
X_test_disc = discretizer.transform(X_test_cont)

X_train_disc_all = np.hstack((X_train_cat, X_train_disc))
X_test_disc_all = np.hstack((X_test_cat, X_test_disc))
```

### 4.4 Modelos mejorando hiperparámetros

**GridSearchCV** es una técnica que realiza una **búsqueda exhaustiva** sobre un conjunto especificado de valores de hiperparámetros. Su objetivo es encontrar la **mejor combinación de hiperparámetros** para un modelo y conjunto de datos determinado, entrenando y evaluando el rendimiento del modelo para cada combinación, y **eligiendo la mejor según la métrica de evaluación definida, en nuestro caso F1-score**.

En cuanto a la **validación**, hemos utilizado un parámetro llamado **cv=5** que hace que cada combinación de hiperparrámetros **se evalúe 5 veces con diferentes particiones de los datos**, y se usa la **media** de esas evaluaciones para decidir cuál combinación es la mejor.

### 4.5 Ensembles

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import StackingClassifier
```

El **StackingClassifier** es una técnica de aprendizaje por ensemble que mejora el rendimiento **combinando las predicciones de múltiples modelos**. Este modelo consta de dos niveles:
- **Modelos base**. Se entrenan con el conjunto de datos original y sus predicciones se usan como características de entrada para el meta-modelo.
- **Meta-modelo**. Es un clasificador final que aprende a **combinar las salidas de los modelos base** para mejorar la precisión.

Nosotros como modelos base hemos empleado los modelos **LogisticRegression, RandomForestClassifier y LGBMClassifier, y como meta-modelo LogisticRegression**. Hemos utilizado esta combinación porque era la que mejores resultados nos daba.

```python
model_rf = RandomForestClassifier(n_estimators=100,class_weight='balanced',max_depth=7,random_state=42)
model_lgb = LGBMClassifier(class_weight='balanced',random_state=42,n_estimators=300,learning_rate=0.05,max_depth=7,reg_lambda=1.0)

meta_model = LogisticRegression(max_iter=1000, class_weight='balanced',random_state=42)

stacking = StackingClassifier(
    estimators=[('lr', meta_model),('rf', model_rf),('lgb', model_lgb)],
    final_estimator=meta_model,
    cv=5,
    n_jobs=-1,
    passthrough=False  
)

stacking.fit(X_train_scaled_onehot, y_train)
y_pred = stacking.predict(X_test_scaled_onehot)
f1 = f1_score(y_test, y_pred)

print(f"F1-score en test: {f1:.4f}")
```

Hemos utilizado el modelo **RandomizedSearchCV**, que es una técnica de ajuste de hiperparámetros que toma **muestras aleatorias de combinaciones de parámetros**. A diferencia de GridSearchCV, que prueba exhaustivamente todas las combinaciones posibles, esta técnica solo prueba un **subconjunto de ellas**, lo que es más **eficiente** cuando el espacio de búsqueda es grande y el modelo costoso.

```python
model_rf = RandomForestClassifier(class_weight='balanced', random_state=42)
model_lgb = LGBMClassifier(class_weight='balanced', random_state=42)

meta_model = LogisticRegression(max_iter=1000, class_weight='balanced', random_state=42)

stacking = StackingClassifier(
    estimators=[('lr', meta_model),('rf', model_rf),('lgb', model_lgb)],
    final_estimator=meta_model,
    cv=5,
    n_jobs=-1,
    passthrough=False
)

param_dist = {
    # RandomForest
    'rf__n_estimators': [100, 200, 300,400],
    'rf__max_depth': [5, 7, 10],
    'rf__min_samples_leaf': [1, 3, 5],

    # LGBM
    'lgb__n_estimators': [100, 200, 300,450],
    'lgb__learning_rate': [0.01, 0.05, 0.1,0.5],
    'lgb__max_depth': [5, 7, 10],
    'lgb__reg_lambda': [0.0, 1.0, 5.0],

    # LogisticRegression 
    'lr__C': [0.1, 1.0, 10.0],

    # Meta-modelo
    'final_estimator__C': [0.1, 1.0, 10.0]
}

search = RandomizedSearchCV(estimator=stacking,param_distributions=param_dist,n_iter=30,cv=5,scoring='f1',n_jobs=-1,verbose=0,random_state=42)

search.fit(X_train_scaled_onehot, y_train)
y_pred = search.predict(X_test_scaled_onehot)
f1 = f1_score(y_test, y_pred)

print(f"F1-score en test: {f1:.4f}")
```
